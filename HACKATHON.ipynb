```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import os

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
```

```python
dataset_path = "/content/dataset/Offroad_Segmentation_Training_Dataset"

train_img_dir = os.path.join(dataset_path, "train/Color_Images")
train_mask_dir = os.path.join(dataset_path, "train/Segmentation")

val_img_dir = os.path.join(dataset_path, "val/Color_Images")
val_mask_dir = os.path.join(dataset_path, "val/Segmentation")

print("Train images:", len(os.listdir(train_img_dir)))
print("Train masks:", len(os.listdir(train_mask_dir)))
```

```python
class OffroadDataset(Dataset):

    def __init__(self, img_dir, mask_dir):
        self.img_dir = img_dir
        self.mask_dir = mask_dir
        self.images = sorted(os.listdir(img_dir))

        self.img_transform = transforms.Compose([
            transforms.Resize((256,256)),
            transforms.ToTensor()
        ])

        self.mask_transform = transforms.Compose([
            transforms.Resize((256,256)),
            transforms.ToTensor()
        ])

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):

        img_name = self.images[idx]

        img_path = os.path.join(self.img_dir, img_name)
        mask_path = os.path.join(self.mask_dir, img_name)

        image = Image.open(img_path).convert("RGB")
        mask = Image.open(mask_path).convert("L")

        image = self.img_transform(image)
        mask = self.mask_transform(mask)

        return image, mask
```

```python
train_dataset = OffroadDataset(train_img_dir, train_mask_dir)
val_dataset = OffroadDataset(val_img_dir, val_mask_dir)

print("Train dataset:", len(train_dataset))
print("Val dataset:", len(val_dataset))
```

```python
train_loader = DataLoader(
    train_dataset,
    batch_size=8,
    shuffle=True,
    num_workers=2
)

val_loader = DataLoader(
    val_dataset,
    batch_size=8,
    shuffle=False
)
```

```python
imgs, masks = next(iter(train_loader))

print("Image shape:", imgs.shape)
print("Mask shape:", masks.shape)
print("Unique mask values:", torch.unique(masks))
```

```python
class SimpleSegNet(nn.Module):

    def __init__(self):
        super().__init__()

        self.encoder = nn.Sequential(
            nn.Conv2d(3,16,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16,32,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(32,16,2,stride=2),
            nn.ReLU(),

            nn.ConvTranspose2d(16,1,2,stride=2)
        )

    def forward(self,x):

        x = self.encoder(x)
        x = self.decoder(x)

        return x
```

```python
model = SimpleSegNet().to(device)

criterion = nn.BCEWithLogitsLoss()

optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
EPOCHS = 20

for epoch in range(EPOCHS):

    model.train()

    total_loss = 0

    for imgs, masks in train_loader:

        imgs = imgs.to(device)

        masks = masks.to(device)

        # convert grayscale mask to binary
        masks = (masks > 0).float()

        optimizer.zero_grad()

        preds = model(imgs)

        if preds.shape != masks.shape:

            preds = F.interpolate(
                preds,
                size=masks.shape[-2:],
                mode='bilinear',
                align_corners=False
            )

        loss = criterion(preds, masks)

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)

    print(f"Epoch {epoch+1}/{EPOCHS} Loss: {avg_loss:.4f}")
```

```python
def compute_iou(pred, mask):

    pred = torch.sigmoid(pred)
    pred = (pred > 0.5).float()

    intersection = (pred * mask).sum()
    union = pred.sum() + mask.sum() - intersection

    return (intersection + 1e-6) / (union + 1e-6)


model.eval()

ious = []

with torch.no_grad():

    for imgs, masks in val_loader:

        imgs = imgs.to(device)
        masks = (masks > 0).float().to(device)

        preds = model(imgs)

        if preds.shape != masks.shape:
            preds = torch.nn.functional.interpolate(
                preds,
                size=masks.shape[-2:],
                mode="bilinear",
                align_corners=False
            )

        iou = compute_iou(preds, masks)

        ious.append(iou.item())

print("Mean IoU:", sum(ious)/len(ious))
```

```python
import torch
import torch.nn as nn

# Redefine the model architecture (SimpleSegNet) since it was not defined
class SimpleSegNet(nn.Module):

    def __init__(self):
        super().__init__()

        self.encoder = nn.Sequential(
            nn.Conv2d(3,16,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16,32,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(32,16,2,stride=2),
            nn.ReLU(),

            nn.ConvTranspose2d(16,1,2,stride=2)
        )

    def forward(self,x):

        x = self.encoder(x)
        x = self.decoder(x)

        return x

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Instantiate the model
model = SimpleSegNet().to(device)

# Load the state dictionary into the model
model.load_state_dict(torch.load(model_path, map_location=device))
model.eval() # Set the model to evaluation mode

print("Model loaded successfully and set to evaluation mode.")
```
